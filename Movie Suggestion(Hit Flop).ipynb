```python
x=df.drop('Status',axis=1)
y=df['Status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
df1=pd.DataFrame({'Actual Status': y_test,'Predcited Status':y_pred})
df1.head(15)
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

```python
def dic_creator(test_keys,test_values): 
     res = {}
     test_keys=list(test_keys)
     test_values=list(test_values)
     for key in test_keys:
         for value in test_values:
             res[key] = value
             test_values.remove(value)
             break 
     return res
```

```python
dic_actor=dic_creator(actor,encode_actor)
dic_actor
```

```python
dic_director=dic_creator(director,encode_director)
dic_director
```

```python
dic_genre=dic_creator(genre,encode_genre)
dic_genre
```

```python
dic_homepage=dic_creator(homepage,encode_homepage)
dic_homepage
```

```python
dic_status=dic_creator(status,encode_status)
dic_status
```

```python
def get_status(val):
     for key, value in dic_status.items():
         if val == value:
             return key
```

```python
def prediction(Actor,Director,Genre,Homepage,Rating):
     a=dic_actor[str(Actor)]
     d=dic_director[Director]
     g=dic_genre[Genre]
     h=dic_homepage[Homepage]
     r=Rating

     ypred=model.predict([[a,d,g,h,r]])
     predi=get_status(ypred)
     return predi
```
